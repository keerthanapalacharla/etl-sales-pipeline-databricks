```python
# Create catalog/schema
spark.sql("CREATE DATABASE IF NOT EXISTS etl_demo")

# Load the raw table
df = spark.table("workspace.default.retail_sales_dataset")

print("Original Columns:")
print(df.columns)

# CLEAN COLUMN NAMES
# Replace spaces, special characters with underscores
clean_cols = [c.lower()
                .replace(" ", "_")
                .replace("/", "_")
                .replace("-", "_")
                .replace("(", "")
                .replace(")", "")
                .replace(".", "_")
                .replace(",", "_")
                for c in df.columns]

# Apply the new columns
df_clean = df.toDF(*clean_cols)

print("Cleaned Columns:")
print(df_clean.columns)

# Register cleaned BRONZE table as managed delta
df_clean.write.mode("overwrite").saveAsTable("etl_demo.bronze_retail_sales")

display(df_clean.limit(10))
# test sync

```
